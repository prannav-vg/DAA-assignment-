```python
import random
import time
import matplotlib.pyplot as plt


# Linear Search Implementation

def linear_search(arr, key):
    for i, val in enumerate(arr):
        if val == key:
            return i
    return -1


# High-Precision Timing Function

def measure_time(func, arr, key, trials=100):
    total_time = 0
    for _ in range(trials):
        start = time.perf_counter()
        func(arr, key)
        end = time.perf_counter()
        total_time += (end - start)
    average_time_ms = (total_time / trials) * 1000  # convert to milliseconds
    return average_time_ms


# Setup

n = 10000  # Size of the input list
data = random.sample(range(1, n * 10), n)  # Unique random integers

# Define keys for each case
key_best = data[0]           # First element (Best Case)
key_average = data[n // 2]   # Middle element (Average Case)
key_worst = -1               # Not in list (Worst Case)


# Measure Execution Times

print("Running performance tests with", n, "elements and 100 trials each...\n")

best_time = measure_time(linear_search, data, key_best)
average_time = measure_time(linear_search, data, key_average)
worst_time = measure_time(linear_search, data, key_worst)


# Display Results

cases = ['Best Case', 'Average Case', 'Worst Case']
times = [best_time, average_time, worst_time]

for case, t in zip(cases, times):
    print(f"{case}: {t:.6f} ms")


# Plotting

plt.figure(figsize=(8, 5))
plt.bar(cases, times, color=['green', 'orange', 'red'])
plt.xlabel('Case Type')
plt.ylabel('Average Time (ms)')
plt.title('Linear Search Time Comparison')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


# Theoretical Analysis

print("\n--- Complexity Analysis ---")
print("Best Case    : O(1)     -> Element found at the first index.")
print("Average Case : O(n/2)  -> Element is somewhere in the middle.")
print("Worst Case   : O(n)     -> Element not present; entire list scanned.")

print("\n--- Observations ---")
print(f"• With input size n = {n}, worst case time is highest.")
print("• Best case is nearly instantaneous (constant time).")
print("• Average case scales with list size, roughly half the work of the worst case.")
print("• As 'n' increases, worst and average cases grow linearly.")
```
