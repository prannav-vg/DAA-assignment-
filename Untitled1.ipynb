```python
import random
import matplotlib.pyplot as plt


# Linear Search with Comparison Count

def linear_search_with_comparisons(arr, key):
    comparisons = 0
    for val in arr:
        comparisons += 1
        if val == key:
            return comparisons
    return comparisons


# Input Sizes to Test

input_sizes = [100, 500, 1000, 5000, 10000]

# Store results
comparisons_best = []
comparisons_avg = []
comparisons_worst = []


# Run Search for Each Size

for n in input_sizes:
    data = list(range(1, n + 1))  # Ordered list (values 1 to n)

    # Best Case: First element
    key_best = data[0]
    comparisons_best.append(linear_search_with_comparisons(data, key_best))

    # Average Case: Middle element
    key_avg = data[n // 2]
    comparisons_avg.append(linear_search_with_comparisons(data, key_avg))

    # Worst Case: Not in list
    key_worst = -1
    comparisons_worst.append(linear_search_with_comparisons(data, key_worst))


# Display Tabular Output

print(f"{'Input Size':<12} {'Best':<10} {'Average':<10} {'Worst':<10}")
print("-" * 42)
for i in range(len(input_sizes)):
    print(f"{input_sizes[i]:<12} {comparisons_best[i]:<10} {comparisons_avg[i]:<10} {comparisons_worst[i]:<10}")


# Plotting

plt.figure(figsize=(10, 6))
plt.plot(input_sizes, comparisons_best, marker='o', label='Best Case (O(1))', color='green')
plt.plot(input_sizes, comparisons_avg, marker='o', label='Average Case (O(n/2))', color='orange')
plt.plot(input_sizes, comparisons_worst, marker='o', label='Worst Case (O(n))', color='red')

plt.title("Linear Search: Number of Comparisons")
plt.xlabel("Input Size (n)")
plt.ylabel("Comparisons")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# Analysis

print("\n--- Theoretical vs Empirical Analysis ---")
print("• Best Case    → 1 comparison (O(1))")
print("• Average Case → ~n/2 comparisons (O(n/2))")
print("• Worst Case   → n comparisons (O(n))")
print("• As n increases, comparisons grow linearly in average & worst case.")
```
